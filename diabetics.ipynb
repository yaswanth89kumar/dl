```python
import pandas as pd
```

```python
data = pd.read_csv("diabetes.csv")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop(['Outcome'],axis=1),data.Outcome,test_size=0.2)
```

```python
import tensorflow as tf
```

```python
#Initialize Sequential model
model = tf.keras.models.Sequential()

#Reshape data from 2D to 1D -> 28x28 to 784
model.add(tf.keras.layers.Reshape((8,),input_shape=(8,)))

#Normalize the data
model.add(tf.keras.layers.BatchNormalization())

#Add Dense Layer 
model.add(tf.keras.layers.Dense(200, activation='relu'))


#Add Dense Layer which provides 10 Outputs after applying softmax - keras declare weights and bias automatically
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

#Comile the model
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train,
          validation_data=(X_test, y_test), 
          epochs=100,
          batch_size=200)
```
